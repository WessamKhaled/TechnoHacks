# Feature Engineering

# Spliting Data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = creditcard.iloc[:,:-1].values
X
```

```python
y = creditcard.iloc[:,-1]
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
svc = SVC()
svc.fit(X_train,y_train)
y_predict_svc = svc.predict(X_test)
print('accuracy_score for Support Vector =',100*accuracy_score(y_predict_svc,y_test),'%')
print('='*68)
print('confusion_matrix for Support Vector\n',confusion_matrix(y_predict_svc,y_test))
print('='*68)
print('classification_report for Support Vector\n',classification_report(y_predict_svc,y_test))
```

```python
reg = LogisticRegression()
reg.fit(X_train,y_train)
y_predict_reg = reg.predict(X_test)
print('accuracy_score for Logistic Regression =',100*accuracy_score(y_predict_reg,y_test),'%')
print('='*68)
print('confusion_matrix for Logistic Regression\n',confusion_matrix(y_predict_reg,y_test))
print('='*68)
print('classification_report for Logistic Regression\n',classification_report(y_predict_reg,y_test))
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train,y_train)
y_predict_tree = tree.predict(X_test)
print('accuracy_score for Decision Tree Classifier =',100*accuracy_score(y_predict_tree,y_test),'%')
print('='*68)
print('confusion_matrix for Decision Tree Classifier\n',confusion_matrix(y_predict_tree,y_test))
print('='*68)
print('classification_report for Decision Tree Classifier\n',classification_report(y_predict_tree,y_test))
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_predict_knn = knn.predict(X_test)
print('accuracy_score for KNeighborsClassifier =',100*accuracy_score(y_predict_knn,y_test),'%')
print('='*68)
print('confusion_matrix for KNeighborsClassifier\n',confusion_matrix(y_predict_knn,y_test))
print('='*68)
print('classification_report for KNeighborsClassifier\n',classification_report(y_predict_knn,y_test))
```

```python
clf = RandomForestClassifier()
clf.fit(X_train,y_train)
y_predict_clf = clf.predict(X_test)
print('accuracy_score for RandomForestClassifier =',100*accuracy_score(y_predict_clf,y_test),'%')
print('='*68)
print('confusion_matrix for RandomForestClassifier\n',confusion_matrix(y_predict_clf,y_test))
print('='*68)
print('classification_report for RandomForestClassifier\n',classification_report(y_predict_clf,y_test))
```
